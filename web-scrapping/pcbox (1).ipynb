```python
from requests_html import HTMLSession
session = HTMLSession()
```

```python
import pandas as pd

max_pagina = 7
datos = []
descripciones_interes = [
    "Capacidad memoria de adaptador gráfico",
    "Tipo de memoria de adaptador gráfico",
    "Cantidad de DisplayPorts",
    "Número de puertos HDMI",
    "Suministro de energía al sistema mínimo",
    "Consumo de energía (max)"
]

for pagina in range(1, max_pagina):
    url = "https://www.pcbox.com/componentes-de-ordenador/tarjetas-graficas?page=" + str(pagina)
    print(f"Pagina {pagina}: {url}")
    
    r = session.get(url)

    linkGrafica = r.html.find("a.vtex-product-summary-2-x-clearLink--main-product-summary")
        
    for link in linkGrafica:
        href = list(link.absolute_links)[0]
        print(f"Enlace del producto: {href}")
        r2 = session.get(href)
        precio = r2.html.find("div.ticnova-commons-components-0-x-price", first=True)
        if precio is not None:
            precioF2  = precio.text.replace("€","")
            precioF3  = precioF2.replace(",",".")
            precioF4 = precioF3.replace(".","").strip()
            precioCentimos = float(precioF4.strip()) * 100
            precioCentimosF =  str(precioCentimos).replace(".","").strip()
            precioFinal =  precioCentimosF[:-1]
            print(precioFinal)
            articulo = r2.html.find("span.vtex-store-components-3-x-productBrand", first=True)
        
            print(f"Marca: {articulo.text.strip()}")
                
            descripcion = r2.html.find("b.vtex-table-description-key-text")
            valor = r2.html.find("td.vtex-table-description-value")

            
            producto_info = {
                "URL" : href,
                "Precio": precioFinal,
                "Articulo": articulo.text.strip()
            }
                
            for i in range(len(descripcion)):
                descripcion_texto = descripcion[i].text.strip()
                valor_texto = valor[i].text.strip()
                if descripcion_texto in descripciones_interes:
                    if descripcion_texto == "Capacidad memoria de adaptador gráfico":
                        producto_info["VRAM"] = valor_texto.replace("GB","").strip()  
                        print(f"{descripcion_texto}: {valor_texto.replace("GB","").strip() }")
                    if descripcion_texto == "Tipo de memoria de adaptador gráfico":
                        producto_info["Tipo VRAM"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Cantidad de DisplayPorts":
                        producto_info["DisplayPort"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Número de puertos HDMI":
                        producto_info["HDMI"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Suministro de energía al sistema mínimo":
                        producto_info["Energia"] = valor_texto.replace("W","").strip()
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Consumo de energía (max)":
                        producto_info["Energia"] = valor_texto.replace("W","").strip()
                        print(f"{descripcion_texto}: {valor_texto}")
            datos.append(producto_info)
        else:
            producto_info = {
                "URL" : href,
                "Precio": "Sin datos",
                "Articulo": "Sin datos",
                "VRAM": "Sin datos",
                "Tipo VRAM": "Sin datos",
                "DisplayPort" : "Sin datos",
                "HDMI": "Sin datos",
                "Energia": "Sin datos"
            }
            datos.append(producto_info)

df =  pd.DataFrame(datos)



```

```python
df.to_csv('tarjetas_graficas.csv', index=False)
```
