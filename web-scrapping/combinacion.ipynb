```python
from requests_html import HTMLSession
session = HTMLSession()
```

```python
datos = []
```

```python
import pandas as pd

max_pagina = 7
descripciones_interes = [
    "Capacidad memoria de adaptador gráfico",
    "Tipo de memoria de adaptador gráfico",
    "Cantidad de DisplayPorts",
    "Número de puertos HDMI",
    "Suministro de energía al sistema mínimo",
    "Consumo de energía (max)"
]

for pagina in range(1, max_pagina):
    url = "https://www.pcbox.com/componentes-de-ordenador/tarjetas-graficas?page=" + str(pagina)
    print(f"Pagina {pagina}: {url}")
    
    r = session.get(url)

    linkGrafica = r.html.find("a.vtex-product-summary-2-x-clearLink--main-product-summary")
        
    for link in linkGrafica:
        href = list(link.absolute_links)[0]
        print(f"Enlace del producto: {href}")
        r2 = session.get(href)
        precio = r2.html.find("div.ticnova-commons-components-0-x-price", first=True)
        if precio is not None:
            precioF2  = precio.text.replace("€","")
            precioF3  = precioF2.replace(",",".")
            precioF4 = precioF3.replace(".","").strip()
            precioFinal =  precioF4 
            print(precioFinal)
            articulo = r2.html.find("span.vtex-store-components-3-x-productBrand", first=True)
        
            print(f"Marca: {articulo.text.strip()}")
                
            descripcion = r2.html.find("b.vtex-table-description-key-text")
            valor = r2.html.find("td.vtex-table-description-value")

            
            producto_info = {
                "URL" : href,
                "Precio": precioFinal,
                "Modelo": articulo.text.strip()
            }
                
            for i in range(len(descripcion)):
                descripcion_texto = descripcion[i].text.strip()
                valor_texto = valor[i].text.strip()
                if descripcion_texto in descripciones_interes:
                    if descripcion_texto == "Capacidad memoria de adaptador gráfico":
                        producto_info["VRAM"] = valor_texto.replace("GB","").strip()  
                        print(f"{descripcion_texto}: {valor_texto.replace("GB","").strip() }")
                    if descripcion_texto == "Tipo de memoria de adaptador gráfico":
                        producto_info["Tipo VRAM"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Cantidad de DisplayPorts":
                        producto_info["DisplayPort"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Número de puertos HDMI":
                        producto_info["HDMI"] = valor_texto
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Suministro de energía al sistema mínimo":
                        producto_info["Energia"] = valor_texto.replace("W","").strip()
                        print(f"{descripcion_texto}: {valor_texto}")
                    if descripcion_texto == "Consumo de energía (max)":
                        producto_info["Energia"] = valor_texto.replace("W","").strip()
                        print(f"{descripcion_texto}: {valor_texto}")
            datos.append(producto_info)
        else:
            producto_info = {
                "URL" : href,
                "Precio": "Sin datos",
                "Modelo": "Sin datos",
                "VRAM": "Sin datos",
                "Tipo VRAM": "Sin datos",
                "DisplayPort" : "Sin datos",
                "HDMI": "Sin datos",
                "Energia": "Sin datos"
            }
            datos.append(producto_info)


```

```python
import re
import pandas as pd

max_pagina = 6

for pagina in range(1, max_pagina):
    url = "https://www.redcomputer.es/tarjetas-graficas-pc-108?page=" + str(pagina)
    print(url)
    r = session.get(url)

    linkGrafica = r.html.find("a.product-thumbnail")
    for link in linkGrafica:
        href = list(link.absolute_links)[0]
        print(href)
        r2 = session.get(href)
        precio = r2.html.find("span.product-price", first=True)
        if  precio is None or precio.text.strip() == "":
            precioFinal = "descatalogado"
        else:
            precioF2  = precio.text.replace("€","")
            precioF3  = precioF2.replace(",",".")
            precioF4 = precioF3.replace(".","").strip()
            precioFinal =  precioF4
            print(precioFinal)
        articulo = r2.html.find("h1.page-title > span", first=True)
        
        producto_info = {
            "URL": href,
            "Precio": precioFinal,
            "Modelo": articulo.text.strip()
        }
        memoria = r2.html.find("div.product-description   li")
        for memor in memoria:
            if any(palabra in memor.text for palabra in ["Memoria:", "Memoria de vídeo", "Memoria.", "Memoria de Video"]):
                separador = "." if "Memoria." in memor.text else ":"
                texto = memor.text.split(separador)
                if "GDDR"  in texto[1]:
                    partes = texto[1].split()
                    vram = partes[0].replace("GB","")        
                    if len(partes) == 3:   
                        tipo_vram = partes[2]
                    else:
                        tipo_vram =  partes[1]
                    print("VRAM:", vram)
                    print("Tipo VRAM:", tipo_vram)
                    producto_info["VRAM"] = vram.strip()
                    producto_info["Tipo VRAM"] = tipo_vram
            if re.search(r'\d+W',memor.text,re.IGNORECASE):
                texto  = memor.text.split(":")
                consumo =  texto[1].replace("W","")
                print("Consumo:", consumo)
                producto_info["Energia"] = consumo.strip()
            if "HDMI" in memor.text:
                texto  = memor.text.split("x")
                if texto[0].strip().isnumeric():
                    print("HDMI", ":" , texto[0])
                    producto_info["HDMI"] = texto[0]
                else:
                    print("HDMI", ":" , 1)
                    producto_info["HDMI"] = 1
            if "displayport" in memor.text.lower():
                texto  = memor.text.split("x")
                if texto[0].strip().isnumeric():
                    print("DisplayPort", ":" , texto[0])
                    producto_info["DisplayPort"] = texto[0]
                else:
                    print("DisplayPort", ":" , 1)
                    producto_info["DisplayPort"] = 1
        datos.append(producto_info)
```

```python
from sklearn.model_selection import train_test_split

X = dfCopia6.drop(['Precio'], axis=1)
y = dfCopia6['Precio']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'R²: {r2}')
print(f'Mean Squared Error (MSE): {mse}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Suponiendo que dfCopia6 es tu DataFrame con las columnas necesarias
X = dfCopia6.drop(['Precio'], axis=1)  # Características
y = dfCopia6['Precio']  # Etiqueta (target)

# Dividir en conjunto de entrenamiento y prueba (30% para prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Inicializar el modelo GradientBoostingRegressor
gb = GradientBoostingRegressor(random_state=0)

# Entrenar el modelo
gb.fit(X_train, y_train)

# Hacer predicciones sobre el conjunto de prueba
y_pred = gb.predict(X_test)

# Calcular el R² y el MSE
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# Imprimir los resultados
print(f'R²: {r2}')
print(f'Mean Squared Error (MSE): {mse}')
```
